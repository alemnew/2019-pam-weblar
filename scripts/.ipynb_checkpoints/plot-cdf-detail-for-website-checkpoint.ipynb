```python
# imports and env't config 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import json
import requests 
from imp import reload 
import seaborn as sns
import Util as util; reload(util)
import warnings
from matplotlib.ticker import FuncFormatter
import matplotlib.ticker as ticker

import time 

import Cdf
import Pmf
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
apr = pd.read_csv('../dataset/monroe_weblar_merged_201804.csv', sep = ',', encoding = "ISO-8859-1")
jul = pd.read_csv('../dataset/monroe_weblar_merged_201807.csv', sep = ',', encoding = "ISO-8859-1")
apr['atf'] = -1
apr['tlsHandshake'] = -1

frames = [apr, jul]
df = pd.concat(frames, ignore_index =  True)

df = df[df['operator'] != 'ice.net']
df = df[df['operator'] != 'Telenor']
df['operator'] = df['operator'].apply(lambda x: util.change_op_cc(x))
df = df[df['url'] != 'https://www.ebay.com/']
df = df[df['url'] != 'https://www.amazon.com/']
df['url'] = df['url'].apply(lambda x: util.get_base_url_name(x))

df = df[df['3sec_rendering_time'] > 0]
df = df[df['10sec_rendering_time'] > 0]
df = df[df['14sec_rendering_time'] > 0]
df = df[df['atf_integral'] > 0]
df = df[df['atf_instant'] > 0]


speed_df = pd.read_csv('../dataset/monroe_weblar_merged_with_speed.csv', sep = ',', encoding = "ISO-8859-1")
speed_df = speed_df[speed_df['gps_speed'] > 0]
mobile_msmt_id = set(speed_df['measurement_id'])
df_se_mobile = df[df['node_id'].isin(util.se_mobile)]
mobile_msmt_id = mobile_msmt_id.union(set(df_se_mobile['measurement_id']))

df['is_mobile'] = df['measurement_id'].apply(lambda x: util.is_from_mobile(x, mobile_msmt_id))
df_mobile = df[df['measurement_id'].isin(mobile_msmt_id)]
df_stationary = df[~df['measurement_id'].isin(mobile_msmt_id)] #isnot in
df_no_mobile = df[df['node_id'].isin(util.no_mobile)]
df_se_mobile = df[df['node_id'].isin(util.se_mobile)]
df_no_stationary = df[df['node_id'].isin(util.no_stationary)]
df_se_stationary = df[df['node_id'].isin(util.se_stationary)]
```

```python
yahoo = df[df['url'] == 'yahoo']
bbc = df[df['url'] == 'bbc']
wiki = df[df['url'] == 'wikipedia']
len(mobile_msmt_id)
```

```python
len(yahoo['node_id'])
```

```python
targets = set(df['url'])
operators = set(df['operator'])
```

## Yahoo details

```python
fig = plt.figure(figsize=(3, 1.5))
plt.rcParams['font.size'] = 12

yticks = np.arange(0,1.1,0.2)

ax1 = fig.add_subplot(111)
legends = []
for o in sorted(operators):
    if o == 'Misc':
        continue
    df_t = yahoo[yahoo['operator'] == o]
    atf = df_t['atf_instant']
    
    cdf_atf = Cdf.MakeCdfFromList(atf)
    
    ax1.plot(cdf_atf.xs, cdf_atf.ps, marker= util.M_O[o], linewidth=2, markersize=1, \
             fillstyle='none', color = util.C_O[o])
#     legends.append(o + '[' + str(len(atf.index)) +']')
    legends.append(o)

    
ax1.set_title("Yahoo", fontsize = 12)

ax1.set_xlabel('ATF  time (s)', fontsize=12)
ax1.set_ylabel('CDF', fontsize=12)
ax1.set_xlim([1000,20000])
ax1.set_xscale('log', basex = 10)
formatter = FuncFormatter(lambda y, _: '{:.16g}'.format(y))
ax1.xaxis.set_major_formatter(formatter)

# boarder in the top and right 
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
ax1.yaxis.set_ticks_position('left')
ax1.xaxis.set_ticks_position('bottom')
ax1.spines['bottom'].set_position(('axes', -0.05))
ax1.spines['left'].set_position(('axes', -0.05))

leg = ax1.legend(legends, 
           fontsize=12, markerscale=1., bbox_to_anchor=(1,1.35), ncol = 1, frameon=True)
for legobj in leg.legendHandles:
    legobj.set_linewidth(1.5)
    
fig.subplots_adjust(hspace = 0.65)
fig.subplots_adjust(wspace = 0.3)

plt.savefig('../gfx/cdf_yahoo_atf_per_operator.pdf', format='pdf', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(4, 3))
plt.rcParams['font.size'] = 12

yticks = np.arange(0,1.1,0.2)

ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

legends = []
for o in sorted(operators):
    if o == 'Misc' or o == 'SE_r':
        continue
    df_t = yahoo[yahoo['operator'] == o]
    atf = df_t['atf_instant']
    
    cdf_atf = Cdf.MakeCdfFromList(atf)
    
    ax1.plot(cdf_atf.xs, cdf_atf.ps, marker= util.M_O[o], linewidth=0.5, markersize=2, \
             fillstyle='none', markevery = 10, color = util.C_O[o])
    
    
    rend = df_t['3sec_rendering_time']
    
    cdf_rend = Cdf.MakeCdfFromList(rend)
    
    ax2.plot(cdf_rend.xs, cdf_rend.ps, marker= util.M_O[o], linewidth=0.5, markersize=2, \
             fillstyle='none', color = util.C_O[o])
    
#     legends.append(o + '[' + str(len(atf.index)) +']')
    legends.append(o)

# add SE roaming data for sake of markevery 
df_t = yahoo[yahoo['operator'] == 'SE_r']
atf = df_t['atf_instant']

cdf_atf = Cdf.MakeCdfFromList(atf)

ax1.plot(cdf_atf.xs, cdf_atf.ps, marker= util.M_O['SE_r'], linewidth=0.5, markersize=2, \
         fillstyle='none', color = util.C_O['SE_r'])


rend = df_t['3sec_rendering_time']

cdf_rend = Cdf.MakeCdfFromList(rend)

ax2.plot(cdf_rend.xs, cdf_rend.ps, marker= util.M_O['SE_r'], linewidth=0.5, markersize=2, \
         fillstyle='none', color = util.C_O['SE_r'])

#     legends.append(o + '[' + str(len(atf.index)) +']')
legends.append('SE_r')

ax1.set_title('ATF$_b$  time', fontsize=12, position=(0.15, 0.59))
ax1.set_ylabel('CDF', fontsize=12)
ax1.set_xlim([1000,20000])
ax1.set_xscale('log', basex = 10)

# ax2.set_xlim([1000,20000])
# ax2.set_xscale('log', basex = 10)

formatter = FuncFormatter(lambda y, _: '{:.16g}'.format(y))
ax1.xaxis.set_major_formatter(formatter)

# boarder in the top and right 
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
ax1.yaxis.set_ticks_position('left')
ax1.xaxis.set_ticks_position('bottom')
ax1.spines['bottom'].set_position(('axes', -0.05))
ax1.spines['left'].set_position(('axes', -0.05))

# y axis tick control
start, end = ax1.get_ylim()
ax1.yaxis.set_ticks(np.arange(start, end, 0.5))
ax1.yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.1f'))
ax2.yaxis.set_ticks(np.arange(start, end, 0.5))
ax2.yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.1f'))

ax2.spines['right'].set_color('none')
ax2.spines['top'].set_color('none')
ax2.yaxis.set_ticks_position('left')
ax2.xaxis.set_ticks_position('bottom')
ax2.spines['bottom'].set_position(('axes', -0.05))
ax2.spines['left'].set_position(('axes', -0.05))
ax2.set_title('ATF$_p$  time', fontsize=12, position=(0.2, 0.59))
ax2.set_xlabel('time (ms)', fontsize=12)


leg = ax2.legend(legends, 
           fontsize=12, markerscale=1., bbox_to_anchor=(1,3.19), ncol = 3, frameon=False)
for legobj in leg.legendHandles:
    legobj.set_linewidth(0.5)
    
fig.subplots_adjust(hspace = 0.3)

plt.savefig('../gfx/cdf_yahoo_atf_per_operator.pdf', format='pdf', bbox_inches='tight')
```
